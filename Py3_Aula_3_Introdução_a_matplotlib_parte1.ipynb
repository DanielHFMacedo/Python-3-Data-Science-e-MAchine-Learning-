```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
empresas = ['Apple','Google','Amazon',
          'Microsoft','Tencent','McDonalds',
          'Visa','Facebook','Alibaba','Luis Vuitton']
```

```python
bilhoes = [947.062, 819.573, 705.646,
           611.460, 214.023, 196.526,
           191.032, 186.421, 169.966,
           124.273]
```

```python
# Já é possível plotar com as variáveis
plt.plot(empresas, bilhoes) # Eixo x e y respectivamente
plt.title('Empresas mais ricas do mundo') # título
plt.xlabel('Nomes das empresas')
plt.ylabel('Bilhões')
```

```python
# Redimensionamento
plt.rcParams['figure.figsize']=(15,8) # Essa regra vai valer para os codigos abaixo apenas

plt.plot(empresas,bilhoes)
plt.title("Empresas mais ricas do mundo")
plt.xlabel('Empresas')
plt.ylabel('Bilhões')
plt.show() # Observe que agora o gráfico está dilatado e dá para ver os nomes das empresas mas as letras estão pequenas.
```

```python
# conhecendo o fontsize
plt.plot(empresas,bilhoes)
plt.xlabel('Empresas',fontsize=20) 
plt.ylabel('Bilhoes',fontsize=20)
plt.title('Empresas mais ricas do mundo',fontsize=25)
plt.show()
```

```python
# Mudando a cor e legendando

plt.plot(empresas,bilhoes, color='green')
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.legend(['Bilhões de dólares'], loc='right', fontsize=15)
plt.show()
```

```python
# aumentar expessura da linha
plt.plot(empresas,bilhoes, color='green', lw=10)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.legend(['Bilhões de dólares'], loc='right', fontsize=15)
plt.show()
```

```python
# adicionar marcador
plt.plot(empresas,bilhoes, color='green', lw=3, marker='o')
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.legend(['Bilhões de dólares'], loc='right', fontsize=15)
plt.show()
```

```python
# adicionar marcador
plt.plot(empresas,bilhoes, color='green', lw=3, marker='o')
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.legend(['Bilhões de dólares'], loc='right', fontsize=15)
plt.grid(True)
plt.show()
```

```python
# adicionar linha de grade
plt.plot(empresas,bilhoes, color='green', lw=3, marker='o')
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.legend(['Bilhões de dólares'], loc='right', fontsize=15)
plt.grid(True)
plt.show()
```

```python
# adicionar linha de grade
plt.plot(empresas,bilhoes, color='green', lw=3, marker='o')
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.legend(['Bilhões de dólares'], loc='high right', fontsize=15)
plt.grid(True)
plt.show()
```

```python
# adicionar linha de grade
plt.bar(empresas,bilhoes, color='green')
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.legend(['Bilhões de dólares'], loc='high right', fontsize=15)
plt.grid(True)
plt.show()
```

```python
#gráfico de barra mais simples
plt.bar(empresas, bilhoes)
```

```python
cores=['red','yellow','green','blue','purple','orange','magenta','white','brown','black']
```

```python
plt.bar(empresas, bilhoes, color = cores, edgecolor='black')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.ylabel('Bilhões de dólares', fontsize=20)
plt.xlabel('Empresas', fontsize=20)
plt.grid(True)
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-ticks')
plt.bar(empresas, bilhoes, color = cores, edgecolor='black')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.ylabel('Bilhões de dólares', fontsize=20)
plt.xlabel('Empresas', fontsize=20)
plt.grid(True)
```

```python
# utilizando barh
plt.style.use('seaborn-ticks')
plt.barh(empresas, bilhoes, color = cores, edgecolor='black')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.ylabel('Bilhões de dólares', fontsize=20)
plt.xlabel('Empresas', fontsize=20)
plt.grid(True)
```

```python
# utilizando scatter
plt.style.use('seaborn-ticks')
plt.scatter(empresas, bilhoes, color = cores, edgecolor='black')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.ylabel('Bilhões de dólares', fontsize=20)
plt.xlabel('Empresas', fontsize=20)
plt.grid(True)
```
